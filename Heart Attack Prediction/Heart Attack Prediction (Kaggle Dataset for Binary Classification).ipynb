# Heart Attack Prediction

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras import models 
from keras import layers
```

```python
df=pd.read_csv(r'C:\Users\kumail\Desktop\FAST\Practising using kaggle datasets\datasets_737503_1278636_heart.csv')
```

```python
df.head()
```

```python
#independent features
inputs_all = df.loc[:, df.columns != 'target']
inputs_all.head()
```

```python
#dependent feature
outputs_all=df.loc[:, df.columns == 'target']
outputs_all.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_all, outputs_all, test_size=0.2, random_state = 44) 
#splits the datasets with 20% dataset reserverd for test data
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
#normalizing the data
mean = x_train.mean(axis=0)
x_train -= mean 
std = x_train.std(axis=0)
x_train /= std
x_test -= mean 
x_test /= std
```

```python
y_train = np.asarray(y_train).astype('float32')
y_test = np.asarray(y_test).astype('float32')
```

```python
x_train=np.asarray(x_train).astype('float32')
x_test=np.asarray(x_test).astype('float32')
```

```python
x_train
```

```python
x_train.shape
```

```python

model = models.Sequential() 
model.add(layers.Dense(9, activation='relu', input_shape=(13,)))
model.add(layers.Dense(9, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid')) #sigmoid to give probablity

model.compile(optimizer='rmsprop',  
              loss='binary_crossentropy', #b/c it is binary classification
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=50, batch_size=10)
```

```python
#testing model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
```

```python
print('\nTest loss: {0:.2f}. Test accuracy: {1:.2f}%'.format(test_loss, test_accuracy*100.))
```
